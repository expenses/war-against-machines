```python
import matplotlib.pyplot as plt
import math
import random
#matplotlib inline
```

```python
rotations = [random.random() * 3 * math.pi for _ in range(500)]

for rotation in rotations:
    rotation + math.radians(45)
    
    x, y = math.cos(rotation), math.sin(rotation)
    
    y *= 0.1
    
    rotation = math.atan2(y, x)
    
    plt.plot([0, math.cos(rotation)], [0, math.sin(rotation)])

plt.show()
```
